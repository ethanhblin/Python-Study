# Concatenate string and type conversion

```python

print("your name is " + "Alex"+ " character.") #Combining strings together and is called concatenate 
#concatenate only works between strings
```

```python
num_char = len(input("what is your name"))
print("your name is " + num_char + " character.")#concatenate failed because num_char is not a string, it's a integer
```

```python
#solution 1 - f string
#takes the input and calculate the lengh of characters and then stores it into num_char
num_char = len(input("what is your name"))
print(f'your name is " {num_char} " character.')
```

```python
num_char = len(input("what is your name"))
print("your name is " + f"{num_char}" + " character.")
```

```python
#solution 2 - type conversion
#str() forces the inside to string  
num_char = len(input("what is your name"))
temp = str(num_char)
print("your name is " + temp + " character.")
```

# Project: Life in weeks
input expected life length in years, print out the equivalant life length in weeks(treat each year 365 days, and 30 days a month)
expected life in years will be converted into weeks

```python
#1 year is 365 days and 365 days is 52 weeks 

expected_life_length = float(input("what is your expected life length in years"))

expected_life_length_in_weeks = round((expected_life_length  * 365) / 7, 1)

print(f'your expected life length of {expected_life_length} years is equal to {expected_life_length_in_weeks} weeks')
```

# Project - life in weeks 2

This project calculates how many weeks a person has left to live based on an expected lifespan.

For example, if Ethan plans to live to 80 years old and has already lived 30 years, the program calculates the remaining 50 years and converts that time into weeks.

The goal is to better understand how much time remains by viewing life in weeks instead of years.

```python
#1 year is 365 days and 365 days is 52 weeks 

print("Welcome to Ethans Life Length Calculator")

expected_life_length = float(input("What is your expected life length (yrs)"))

expected_life_length_in_weeks = round((expected_life_length * 365) / 7, 1)

age = float(input("How old are you?"))

age_in_weeks = round((age *365) / 7, 1)

remaining_weeks_left = expected_life_length_in_weeks - age_in_weeks

print(f'your remaining life length will be {remaining_weeks_left} weeks left')




```
